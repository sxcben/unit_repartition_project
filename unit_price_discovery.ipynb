```python
total_price = 4000
nber_of_unit = 4
price_per_unit = total_price / nber_of_unit
list_of_prices = dict()
for i in range(1, nber_of_unit + 1):
    list_of_prices[f"unit_{i}"] = price_per_unit
unit_people = {
    "Karim": "unit_1",
    "Hassan": "unit_2",
    "Benjamin": "unit_3",
    "Last": "unit_4",
}
state_people = {
    "Karim": False,
    "Hassan": False,
    "Benjamin": False,
    "Last": False,
}

while state_people["Karim"] is False:
    unit_karim = unit_people["Karim"]
    price_karim = list_of_prices.get(unit_karim)
    print(price_karim)
    answer = input("Karim with which unit do you wanna switch? ")
    name_person = unit_people.get(f"unit_{answer}")
    get_price = list_of_prices.get(f"unit_{answer}")
    price_proposed = int(input(f"Karim, you wanna switch with unit_{answer} which is worth {get_price}. How much are you willing to pay? "))
    difference = price_proposed - price_karim
    ## Now ask the proprietor of the unit if he agrees to switch
    answer_proprietor = input(f"Proprietor of unit_{answer}, do you agree to switch with Karim for {price_proposed} and get Karim {price_karim - difference} for ? (yes/no) ")
    if answer_proprietor.lower() == "yes":
        state_people["Karim"] = True
        state_people[name_person] = True
        unit_people["Karim"] = f"unit_{answer}"
        unit_people[name_person] = unit_karim
        list_of_prices[f"unit_{answer}"] = price_karim
        list_of_prices[unit_karim] = get_price
print("Karim has successfully switched units!")
```
